```python
#LIBS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dfinitions of initial data
np.random.seed(100)
#Months to analyze
cols = ['Janeiro', 'Março', 'Maio', 'Julho', 'Agosto', 'Outubro', 'Novembro']
#Setting the number of lines
lin = np.arange(1, 32)
#Set teh data
dados = np.random.randint(492, 2050, size=(31,7))
#Dataframe creation
df = pd.DataFrame(data=dados, index = lin, columns = cols)
df
```

## Analysis of the total sales for the month of January

```python
#Lines
x = df.index
```

```python
#Columns
y_january = df['Janeiro']
```

```python
#Graphic
plt.plot(x,y_january)
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

## Comparison of total sales from January to October

```python
#Columns January and October from the Datafram
y_comp = df[['Janeiro', 'Outubro']]
```

```python
#Graphic
plt.plot(x,y_comp)
#Graphic formatation
plt.title("Comparison of January from October Sales")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

## Color Change

```python
#Graphic
plt.plot(x,y_january, color = 'red')
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

## Line Style

```python
#Graphic Dotted
plt.plot(x,y_january, color = 'g', linestyle=':')
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

```python
#Graphic Dashed
plt.plot(x,y_january, color = 'b', linestyle='--')
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

```python
#Graphic Continuous
plt.plot(x,y_january, color = 'gray', linestyle='-')
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

```python
#Graphic Dashed-Dotted
plt.plot(x,y_january, color = 'purple', linestyle='-.')
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

## Marker

```python
#Graphic
plt.plot(x,y_january, color = 'purple', linestyle='-.', marker='^')
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```

## Full Formatation

```python
#Graphic Dashed-Dotted
plt.plot(x,y_january, '+k--')
#Graphic formatation
plt.title("Sales by Day")
plt.xlabel("Month Days")
plt.ylabel("Value (R$)")
```
